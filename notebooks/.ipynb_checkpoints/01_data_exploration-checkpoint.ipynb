```python
df['num_cuisines'] = df['cuisine'].apply(
    lambda x: len(x.split(',')) if isinstance(x, str) else 1
)
```

```python
df['adjusted_demand'] = df['demand'] / df['num_cuisines']
```

```python
df['cuisine'] = df['cuisine'].str.split(',')
df = df.explode('cuisine')
df['cuisine'] = df['cuisine'].str.strip()
```

```python
combo = df.groupby(['location', 'cuisine']).agg({
    'restaurant_id': 'count',
    'adjusted_demand': 'sum',
    'rating': 'mean',
    'price': 'mean'
}).rename(columns={
    'restaurant_id': 'competition',
    'adjusted_demand': 'demand'
}).reset_index()
```

```python
combo['opportunity_score'] = (
    combo['demand'] * combo['rating']
) / (combo['competition'] + 5)
```

```python
combo['normalized_demand'] = (
    combo['demand'] / combo['demand'].max()
) * 100
```

```python
combo['opportunity_score'] = (
    combo['normalized_demand'] * combo['rating']
) / (combo['competition'] + 5)
```

```python
best_per_location = combo.sort_values(
    ['location', 'opportunity_score'], 
    ascending=[True, False]
).groupby('location').head(3)
```

```python
# 1. Create the styled object
styled_report = best_per_location.style \
    .format({
        'rating': '{:.1f} â',           
        'price': 'â¹{:,.0f}',           
        'opportunity_score': '{:,.2f}', 
        'demand': '{:,.0f}'             
    }) \
    .background_gradient(
        subset=['opportunity_score'], 
        cmap='Greens'                   
    ) \
    .background_gradient(
        subset=['competition'], 
        cmap='Reds' 
    ) \
    .set_properties(**{
        'text-align': 'left',
        'font-family': 'Arial',
        'border': '1px solid lightgrey'
    }) \
    .set_caption("Strategic Growth Map: Top 3 Cuisines per Location")

# 2. Display it
styled_report
```
